```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.eval()  # Set to evaluation mode
```

```python
import requests

url = "https://www.gutenberg.org/files/1342/1342-0.txt"
response = requests.get(url)
text = response.text
```

```python
start = text.find("Chapter 1")
end = text.find("End of the Project Gutenberg")
clean_text = text[start:end]
```

```python
def generate_text(prompt, max_length=50, temperature=1.0, top_k=50, top_p=0.9):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print(generate_text("Elizabeth Bennet said"))
```

```python
print(generate_text("Mr. Darcy", max_length=100))
```

```python
print(generate_text("The party arrived", temperature=0.7))
print(generate_text("The party arrived", temperature=1.5))
```

```python
print(generate_text("It is a truth universally", top_k=30, top_p=0.8))
```

```python
print(generate_text("In the 19th century England, a young lady"))
```

```python
print(generate_text("Always remember that"))
```

### Text Generation with GPT-2 using *Pride and Prejudice*

## Introduction  
Generative AI refers to artificial intelligence systems capable of generating new content such as text, images, music, and more. These models learn patterns in data and use probabilistic methods to generate creative or coherent outputs. One popular example is OpenAI’s GPT, which uses deep learning to generate human-like text based on a prompt.

## GPT Architecture  
GPT (Generative Pre-trained Transformer) is based on the Transformer decoder architecture. It uses self-attention mechanisms to process and relate words in a sequence.  
- **Tokenization:** Breaks input into smaller tokens (subwords).  
- **Training:** Predicts the next word given previous words (causal language modeling).  
- **Generation:** Produces new text by sampling one token at a time based on probability distributions.

## Dataset Description  
The dataset used is *Pride and Prejudice* by Jane Austen, sourced from [Project Gutenberg](https://www.gutenberg.org/ebooks/1342). It is a public domain English novel rich in natural language and narrative structure, ideal for testing text generation models.

## Ethical Considerations  
Generative AI raises concerns such as:  
- **Bias** in output based on training data  
- **Misinformation** through plausible but false content  
- **Plagiarism** when using large language models without attribution  
Mitigation strategies include dataset auditing, transparency, and responsible model deployment.
